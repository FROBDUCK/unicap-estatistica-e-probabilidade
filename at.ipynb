```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

# Load the Excel file
file_path = 'base alunos.xlsx'
data = pd.read_excel(file_path)

# Display the first few rows of the dataset to understand its structure
data.head()
# Filtrar as variáveis de interesse
data_filtered = data[['Nota_Teste', 'TipoEsc']]

# Análise descritiva das notas segmentada por tipo de escola
descriptive_stats = data_filtered.groupby('TipoEsc').describe()

# Gráfico Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='TipoEsc', y='Nota_Teste', data=data_filtered)
plt.title('Distribuição das Notas')
plt.xlabel('Tipo de Escola')
plt.ylabel('Nota do Teste')
plt.show()

descriptive_stats

# Separar as notas por tipo de escola
notas_privadas = data_filtered[data_filtered['TipoEsc'] == 'priv']['Nota_Teste']
notas_publicas = data_filtered[data_filtered['TipoEsc'] == 'pub']['Nota_Teste']

# Realizar o teste T
t_stat, p_value = stats.ttest_ind(notas_privadas, notas_publicas)

t_stat, p_value

```
